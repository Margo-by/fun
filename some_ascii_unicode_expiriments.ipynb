```python
# Вывод таблицы ASCII (первые 128 символов)
print("ASCII Table:")
print("{:<5} {:<10} {:<10}".format("Code", "Char", "Description"))  # Заголовок

for i in range(128):
    char = chr(i)  # Получаем символ по коду
    # Специальная обработка для символов, которые не видны
    if i < 32 or i == 127:
        description = f"Control: {i}"
        char = repr(char)  # Показываем символ как строку (например, '\n' для новой строки)
    else:
        description = "Printable"
    print("{:<5} {:<10} {:<10}".format(i, char, description))
```

```python
# Вывод первых символов Unicode (от 0 до 127)
print("\nUnicode Table (0-200):")
print("{:<5} {:<10} {:<10}".format("Code", "Char", "Description"))  # Заголовок

for i in range(201):
    char = chr(i)  # Получаем символ по коду
    # Специальная обработка для символов, которые не видны
    if i < 32 or i >= 127:
        description = f"Control: {i}"
        char = repr(char)  # Показываем символ как строку (например, '\n' для новой строки)
    else:
        description = "Printable"
    print("{:<5} {:<10} {:<10}".format(i, char, description))
```
